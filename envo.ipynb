```python
import copy
import sys

from rdflib import Graph
import pandas as pd
from wikidataintegrator import wdi_core, wdi_login
```

```python
envoRDF = Graph()
envoRDF.parse("https://raw.githubusercontent.com/EnvironmentOntology/envo/master/envo.owl",format="xml")
```

```python
termList = pd.DataFrame(columns=["term", "subClassOf", "label", "synonyms", "wikipedia", "wikidata" ])
termList
```

```python
query = """
   SELECT ?term ?label ?subClassOf ?wikipedia  WHERE {
       ?term rdfs:label ?label .
       ?term oboInOwl:hasDbXref ?wikipedia .
        FILTER (regex(str(?wikipedia), "wikipedia", "i"))
   }

"""
for row in envoRDF.query(query):
    term = row["term"]
    subCLassOf = row["subClassOf"]
    label = row["label"]
    wikipedia = row["wikipedia"]
    synonyms = None
    termList.loc[len(termList)] = [term, subCLassOf, label, synonyms, wikipedia, None]
```

```python
from wikidataintegrator import wdi_core, wdi_login
for index, row in termList.iterrows():
    #print(row["wikipedia"])
    query = f"""
        SELECT * WHERE {{
           <{row["wikipedia"]}> schema:about ?qid .
        }}
    """
    results = wdi_core.WDFunctionsEngine.execute_sparql_query(query)

    print(len(results["results"]["bindings"]))
    if len(results["results"]["bindings"])>0:
        termList.loc[index]["wikidata"] = results["results"]["bindings"][0]["qid"]["value"]
```
